```python
import requests
import pandas as pd
from bs4 import BeautifulSoup 
from openpyxl import Workbook
```

```python
wb = Workbook()
ws = wb.active
```

```python
res = requests.get("https://finance.naver.com/sise/sise_quant.naver")
print(res.status_code)
```

```python
soup = BeautifulSoup(res.content.decode("euc-kr", "replace"), 'html.parser')
print(soup)
```

```python
soup_table = soup.select_one('table.type_2')
print(soup_table)
```

```python
table_rows = soup_table.select("tr")
print(table_rows)
```

```python
# for row in table_rows:
#     print(row.select("td"))

for table_row in table_rows:

    if len(table_row) == 1:
        continue

    stock_datas = []

    for table_data in table_row:
        stock_data = table_data.get_text().strip()
        if stock_data != "":
            stock_datas.append(stock_data)

    ws.append(stock_datas)

wb.save("top_trading.xlsx")
```

```python
soup_table = soup.select_one("table.type_2")
tables = pd.read_html(str(soup_table))
print(tables)

```

```python
table = tables[0]
print(table)
```

```python
table["PBR"] = table.PER * table.ROE / 100
```

```python
stock_datas = table[(table.PER > 0) & (table.PER < 10) & (table.PBR < 1)]
```
